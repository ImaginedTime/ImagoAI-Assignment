# Exploratory Data Analysis and PCA for Vomitoxin Prediction

This notebook analyzes hyperspectral imaging data to predict vomitoxin levels in corn samples. We'll perform:
1. Data Loading and Initial EDA
2. Feature Distribution Analysis
3. Dimensionality Reduction using PCA
4. Model Development and Evaluation

```python
# Separate features and target
X = df.drop(['vomitoxin_ppb'], axis=1)
y = df['vomitoxin_ppb']

print("\nFeature Statistics:")
print(X.describe().round(3))
```

```python
# Identify spectral columns (assuming all columns except target and log-target are spectral)
spectral_cols = [col for col in df.columns if col not in ['vomitoxin_ppb', 'log_vomitoxin_ppb', 'hsi_id']]

# Average spectral reflectance plot

avg_spectrum = df[spectral_cols].mean()
plt.plot(avg_spectrum.values)
plt.title("Average Spectral Reflectance")
plt.xlabel("Wavelength Band Index")
plt.ylabel("Reflectance")
plt.show()

# Correlation heatmap of spectral bands
corr_matrix = df[spectral_cols].corr()
plt.imshow(corr_matrix, aspect='auto')
plt.colorbar()
plt.title("Correlation Heatmap of Spectral Bands")
plt.show()

# Heatmap of reflectance values across samples
plt.imshow(df[spectral_cols].values, aspect='auto')
plt.title("Heatmap of Reflectance Values (Samples x Bands)")
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Original distribution
plt.subplot(1, 2, 1)
sns.histplot(y, bins=30)
plt.title('Distribution of Vomitoxin Levels')
plt.xlabel('Vomitoxin (ppb)')

# Log-transformed distribution
plt.subplot(1, 2, 2)
sns.histplot(np.log1p(y), bins=30)
plt.title('Log-transformed Vomitoxin Levels')
plt.xlabel('Log(Vomitoxin + 1) (ppb)')

plt.tight_layout()
plt.show()

print("\nVomitoxin Statistics:")
print(y.describe().round(2))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Calculate explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# Plot explained variance
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, 'bo-')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')

plt.subplot(1, 2, 2)
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, 'ro-')
plt.axhline(y=0.95, color='k', linestyle='--')
plt.title('Cumulative Explained Variance Ratio')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')

plt.tight_layout()
plt.show()

# Find number of components for 95% variance
n_components_95 = np.argmax(cumulative_variance_ratio >= 0.95) + 1
print(f"Number of components needed for 95% variance: {n_components_95}")
```

```python
# Reduce dimensions using PCA
pca_final = PCA(n_components=n_components_95)
X_reduced = pca_final.fit_transform(X_scaled)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_reduced, y, test_size=0.2, random_state=42
)
```

```python
# Train Random Forest model
rf_model = RandomForestRegressor(n_estimators=600, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Calculate metrics
r2 = r2_score(y_test, y_pred)
adjusted_r2 = 1 - (1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"RÂ² Score: {r2:.3f}")
print(f"Adjusted RÂ² Score: {adjusted_r2:.3f}")
print(f"RMSE: {rmse:.2f} ppb")
```

```python
plt.figure(figsize=(12, 5))

# Actual vs Predicted
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Vomitoxin (ppb)')
plt.ylabel('Predicted Vomitoxin (ppb)')
plt.title('Actual vs Predicted Values')

# Residuals
plt.subplot(1, 2, 2)
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Vomitoxin (ppb)')
plt.ylabel('Residuals')
plt.title('Residual Plot')

plt.tight_layout()
plt.show()
```

## Conclusions

1. The spectral data was successfully reduced from 448 dimensions to a smaller set while preserving 95% of the variance.
2. The target variable (vomitoxin_ppb) shows a right-skewed distribution, suggesting log transformation was appropriate.
3. The Random Forest model achieved reasonable prediction accuracy, though there's room for improvement.
4. The residual plot can help identify any systematic prediction errors.
